```python
import requests
import geopandas as gpd
from shapely.geometry import Point

# Define the API endpoint and your API key
api_endpoint = "https://api.cyclestreets.net/v2/advocacydata.modalfilters"
api_key = 

# Define the bounding box for the new area
bbox = "-1.7774,54.9404,-1.5202,55.0815"

# Construct the request URL
url = f"{api_endpoint}?bbox={bbox}&key={api_key}"

# Make the API call
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()
    
    # Extract features
    features = data['features']
    
    # Extract properties and coordinates
    properties = [feature['properties'] for feature in features]
    coordinates = [feature['geometry']['coordinates'] for feature in features]
    
    # Create GeoDataFrame
    geometry = [Point(coord) for coord in coordinates]
    gdf = gpd.GeoDataFrame(properties, geometry=geometry, crs="EPSG:4326")
    
    # Reproject to CRS 27700
    gdf = gdf.to_crs(epsg=27700)
    
    # Rename GeoDataFrame to CS_Filters
    gdf.name = "CS_Filters"
    
    # Display GeoDataFrame
    print(gdf.head())
    
    # Save GeoDataFrame to GeoPackage file
    output_path = r"C:\Users\b8008458\Documents\scratch_space\CS_Filters.gpkg"
    gdf.to_file(output_path, driver='GPKG')
    print(f"GeoDataFrame saved to: {output_path}")
else:
    print("Error:", response.json())
```
