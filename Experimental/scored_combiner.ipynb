```python
import os
import geopandas as gpd
import pandas as pd

# Define the path to the directory containing the GeoPackages
folder_path = r'C:\Users\b8008458\Documents\scratch_space\linux_outputs\outputs\scored_neighbourhoods\comb'

# List to hold each GeoDataFrame
gdf_list = []

# Loop over each file in the directory
for file_name in os.listdir(folder_path):
    if file_name.endswith(".gpkg"):  # Check if it's a GeoPackage file
        file_path = os.path.join(folder_path, file_name)
        
        # Read the GeoPackage into a GeoDataFrame
        gdf = gpd.read_file(file_path)
        
        # Append the GeoDataFrame to the list
        gdf_list.append(gdf)

# Concatenate all the GeoDataFrames into one
combined_gdf = pd.concat(gdf_list, ignore_index=True)

# Save the combined GeoDataFrame as a new GeoPackage (optional)
output_path = os.path.join(folder_path, 'combined.gpkg')
combined_gdf.to_file(output_path, driver="GPKG")

print(f"Combined GeoPackage saved to {output_path}")
```
