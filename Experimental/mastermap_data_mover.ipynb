```python
import os
import shutil

# Define the source directory
source_dir = r'C:\Users\b8008458\OneDrive - Newcastle University\2022 to 2023\PhD\GIS Dataset\processed_mastermap_roads'

# Define the destination directory
destination_dir = r'C:\Users\b8008458\OneDrive - Newcastle University\2022 to 2023\PhD\GIS Dataset\processed_mastermap_roads\New folder'

# Function to move files from subdirectories to the target folder
def move_files(source_dir, destination_dir):
    # Iterate through each subdirectory in the source directory
    for root, dirs, files in os.walk(source_dir):
        # Iterate through each file in the current subdirectory
        for file in files:
            # Construct the full path of the file
            file_path = os.path.join(root, file)
            # Check if the file is not in the main directory
            if os.path.dirname(file_path) != source_dir:
                # Construct the destination path
                dest_file_path = os.path.join(destination_dir, file)
                # Check if a file with the same name already exists in the destination directory
                if not os.path.exists(dest_file_path):
                    # Move the file to the destination directory
                    shutil.move(file_path, dest_file_path)
                    print(f"Moved {file_path} to {dest_file_path}")
                else:
                    print(f"File {file} already exists in {destination_dir}. Skipping...")

# Call the function to move files
move_files(source_dir, destination_dir)
```
